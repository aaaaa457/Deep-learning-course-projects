```python
import os
from moviepy import VideoFileClip  

def convert_mp4_to_wav(video_dir, audio_dir):
    # 确保目标音频文件夹存在，如果不存在则创建
    if not os.path.exists(audio_dir):
        os.makedirs(audio_dir)

    # 遍历指定目录下的所有文件
    for filename in os.listdir(video_dir):
        if filename.endswith(".mp4"):
            mp4_file = os.path.join(video_dir, filename)
            # 创建新的wav文件名，并将其放置在目标音频文件夹中
            wav_filename = os.path.splitext(filename)[0] + ".wav"
            wav_file = os.path.join(audio_dir, wav_filename)
            
            # 检查目标文件夹中是否存在同名的.wav文件
            if os.path.exists(wav_file):
                print(f"Skipping conversion of {mp4_file} as {wav_file} already exists.")
                continue  # 如果文件已存在，则跳过此次循环
            
            print(f"Converting {mp4_file} to {wav_file}")
            try:
                # 创建一个VideoFileClip对象并提取音频
                video = VideoFileClip(mp4_file)
                video.audio.write_audiofile(wav_file)
                print(f"Successfully converted {mp4_file}")
            except Exception as e:
                print(f"Failed to convert {mp4_file}: {e}")

video_directory = "./data/VGGSound/video"
audio_directory = "./data/VGGSound/audio"
convert_mp4_to_wav(video_directory, audio_directory)
```

数据集相应的vggsound2.csv文件

```python
import pandas as pd
import os
import re
import numpy as np

# 定义路径
csv_path = './data/VGGSound/vggsound.csv'
video_dir = './data/VGGSound/video'
output_csv_path = './data/VGGSound/vggsound2.csv'

# 读取CSV文件
df = pd.read_csv(csv_path)

# 准备一个空列表来存储匹配的行
matching_rows = []

# 获取视频文件列表
for filename in os.listdir(video_dir):
    if filename.endswith('.mp4'):
        # 提取ytid和start_time_from_filename
        ytid = filename[:11]
        # 假设文件名为：{ytid}{start_time}.mp4，使用正则表达式提取数字部分
        match = re.search(r'(\d+)\.mp4$', filename)
        if match:
            start_time_from_filename = int(match.group(1))
            
            # 查找DataFrame中对应的行
            matching_row = df[(df['ytid'] == ytid) & (df['start_time'] == start_time_from_filename)]
            if not matching_row.empty:
                matching_rows.append(matching_row)

# 将所有匹配的行连接成一个新的DataFrame
filtered_df = pd.concat(matching_rows, ignore_index=True)

# 随机选取25%的数据索引
random_indices = np.random.choice(filtered_df.index, size=int(len(filtered_df) * 0.25), replace=False)

# 修改这25%数据的'set'列为'test'
filtered_df.loc[random_indices, 'set'] = 'test'

# 将筛选后的DataFrame保存到新的CSV文件
filtered_df.to_csv(output_csv_path, index=False)
```
